# 1 Acquire Journal Description

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
def get_journal_description(url, journal_name):

    options = Options()
    options.add_argument('--headless')

    driver = webdriver.Chrome(executable_path='C:\ChromeDriver\chromedriver.exe', options=options)

    driver.get(url)

    elements = driver.find_elements_by_xpath("//div[@class='widget-body body body-none  body-compact-all']")

    aim_text=''

    for idx, div in enumerate(elements):
        temp_text = div.text.replace('&', 'and').lower()
        if ((journal_name in temp_text) | ('the journal' in temp_text)) & ('aims and scope'in temp_text):
            #print(idx)
            #print(temp_text)
            aim_text = temp_text
            #break

    driver.quit()

    return aim_text


```

```python
journal_df['Aim']=''


for index, row in journal_df.iterrows():
    journal_id = row['title_id']
    journal_name = row['publication_title']
    url = aim_link+journal_id
    aim_text = get_journal_description(url, journal_name)
    row['Aim'] = aim_text
    if aim_text == '':
        print(index)

journal_df.describe()
journal_df

    
```
