# 1 text manipulation

```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download("stopwords")
new_stop_words=stopwords.words("english")
```

```python
def remove_spec_char(df, col_name):
    df[col_name] = df[col_name].str.replace("n't",'not').str.strip()
    df[col_name] = df[col_name].str.replace(r'[0-9]+', "").str.strip()
    df[col_name] = df[col_name].str.replace(r"[^a-zA-Z ]+", " ").str.strip()

    #new_text= re.sub("n't",'not', df)
    #new_text = re.sub(r'[0-9]+', '', df)
    #new_text= re.sub(r"[^a-zA-Z0-9]+", ' ', df)
    return df
```

```python
df_paper_journal=remove_spec_char(df_paper_journal, "paper_title")
df_paper_journal=remove_spec_char(df_paper_journal, "keywords")
df_paper_journal.head(10)
```
